Ahora verifiquemos los valores mínimos y máximos de `year_of_release` de todos los juegos con un pivot_table. La idea es que el valor mínimo y máximo sean los mismos. 

```python
# pivot_table
year_min_max = pd.pivot_table(games, 
                              values = "year_of_release", 
                              index = "name", 
                              aggfunc = ["min", "max"])
year_min_max
```

Verificamos los nombres de las columnas `year_min_max`

```python
# nombres columnas
year_min_max.columns
```

Definimos una función para identificar que si el año mínimo y máximo son iguales, obtenga un valor, sino, que se reemplace con 0 (esto serán todos los juegos que tuvieron distintos años de lanzamiento). 

```python
# función 
def min_max(row):
    min_year = row[('min', 'year_of_release')]
    max_year = row[('max', 'year_of_release')]
    
    if min_year == max_year: 
        return min_year
    else:
        return 0
```

```python
# aplicación a la función 
year_min_max["year_of_release"] = year_min_max.apply(min_max, axis=1)
```

```python
#verificación
year_min_max.head()
```

```python
# quitar todos los que son igual a 0
year_min_max = year_min_max[year_min_max["year_of_release"] != 0]
year_min_max
```

Para poder concluir mejor sobre el sistema ESRB es mejor entender qué significa cada tipo de letra:

- `M`: Mature 17+

- `T`: Teen (suitable for ages 13 and over)

## Prueba de hipótesis

### Calificaciones promedio de usuarios para plataformas `XOne` y `PC` son las mismas

Para analizar si las calificaciones promedio de usuarios para plataformas `XOne` y `PC` son las mismas se utilizará la verificación de hipótesis sobre la igualdad de las medias de dos poblaciones. 

Antes de realizar esto, procedemos a filtrar la tabla de games limpia, junto con las calificaciones de los usuarios solo para los juegos de `XOne` y `PC`. 

```python
# filtro de PC y XOne user_score
xone_user_score = games.query('platform == "XOne"')["user_score"]
pc_user_score = games.query('platform == "PC"')["user_score"]
```

Verificamos cuántos datos hay en cada `XOne` y `PC`: 

```python
# total datos xbox one
len(xone_user_score)
```

```python
# total datos pc
len(pc_user_score)
```

Tomando en cuenta que la cantidad de datos en cada plataforma difiere, se debe tener una misma cantidad de datos para poder analizar y estudiar la prueba de hipótesis. Para esto, se decide tomar una muestra aleatoria de 150 datos para `XOne` y `PC`, sin tomar en cuenta los valores nulos. 

```python
# 150 datos aleatorios xbox 
xone_150_data = xone_user_score.dropna().sample(n=150, random_state=1)
xone_150_data
```

```python
# 150 datos aleatorios pc
pc_150_data = pc_user_score.dropna().sample(n=150, random_state=1)
pc_150_data
```

Ya con los datos listos, establecemos nuestras hipótesis de la siguiente manera: 

Hipótesis nula: 
- Las calificaciones promedio de los usuarios para las plataformas `XOne` y `PC` son las mismas. 

Hipótesis alternativa: 
- Las calificaciones promedio de los usuarios para las plataformas `XOne` y `PC` difieren. 

A partir de esto, se calculará el valor p según la igualdad de las medias de dos poblaciones con la prueba T aplicando `st.ttest_ind`, con un alpha de 0.05. 

```python
# comprobación de hipótesis
alpha = 0.05 

results = st.ttest_ind(xone_150_data, pc_150_data, equal_var = False)

print('p-value:', results.pvalue)

if results.pvalue < alpha:
    print("Rechazamos la hipótesis nula")
else:
    print("No rechazamos la hipótesis nula")
```

El valor p calculado es menor que 0.05, por lo tanto se rechaza la hipótesis nula. En este caso hay suficiente información estadística para decir que las calificaciones promedio de los usuarios de `XOne` y `PC` son diferentes. 

### Calificaciones promedio de usuarios para género de acción y deportes son diferentes

En este caso queremos comprobar si las calificaciones promedio de usuarios para los videojuegos de género de acción y deportes son diferentes. El procedimiento que se realizó en el 6.1 es similar a este, por lo tanto se procede a realizar los mismos pasos:

```python
# filtro de género acción y deportes user_score
action_user_score = games.query('genre == "Action"')["user_score"]
sports_user_score = games.query('genre == "Sports"')["user_score"]
```

```python
# longitud de cada grupo de datos
print(len(action_user_score))
print(len(sports_user_score))
```

De igual forma, tenemos datos diferentes en cada set de datos, por lo tanto se igual obteniendo de forma aleatoria una muestra de 1500 datos, sin tomar en cuenta los valores nulos. 

```python
# 1500 datos aleatorios action 
action_1500_data = action_user_score.dropna().sample(n=1500, random_state=1)
action_1500_data
```

```python
# 1500 datos aleatorios sports 
sports_1500_data = sports_user_score.dropna().sample(n=1500, random_state=1)
sports_1500_data
```

Ya con los datos listos, establecemos nuestras hipótesis de la siguiente manera: 

Hipótesis nula: 
- Las calificaciones promedio de los usuarios para las géneros de `Action` y `Sports` son las mismas. 

Hipótesis alternativa: 
- Las calificaciones promedio de los usuarios para las géneros de `Action` y `Sports` difieren. 

A partir de esto, se calculará el valor p según la igualdad de las medias de dos poblaciones con la prueba T aplicando `st.ttest_ind`, con un alpha de 0.05. 

```python
# comprobación de hipótesis
alpha = 0.05 

results = st.ttest_ind(action_1500_data, sports_1500_data, equal_var = False)

print('p-value:', results.pvalue)

if results.pvalue < alpha:
    print("Rechazamos la hipótesis nula")
else:
    print("No rechazamos la hipótesis nula")
```

Como el valor P es menor que 0.05, rechazamos la hipótesis nula, por lo tanto hay suficiente información estadística para decir que los videojuegos de género de acción y deporte si tienen distintas calificaciones promedio entre los usuarios.
